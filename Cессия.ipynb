# Тренируем регриссионную модель

```python
df.columns
```

```python
X = df[['deductions_of_subsoil_users_for_the_socio-economic_development']]
y = df['deductions_of_subsoil_users_for_training']
```

# Используем model_selection.train_test_split от sklearn, чтобы разделить данные на наборы для обучения и тестирования. Установим test_size=0.3 и random_state=101

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

# Обучение модели

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

# Создаем экземпляр модели LinearRegression() с именем lr

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

# (Train/fit) Обучение/подгонка lm на тренировочных данных.

```python
lr.fit(X_train, y_train)
```

# Распечатаем коэффициенты модели

```python
lr.coef_
```

# lr.predict() для предсказания набора данных X_test.


```python
predictions = lr.predict(X_test)
```

# Создадим scatterplot

```python
plt.scatter(y_test, predictions);
```

# Рассчитаем среднюю абсолютную ошибку, среднеквадратичную ошибку и среднеквадратическое отклонение (Mean Absolute Error, Mean Squared Error, and the Root Mean Squared Error).

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, explained_variance_score
print('MAE : {}'.format(mean_absolute_error(y_test, predictions)))
print('MSE : {}'.format(mean_squared_error(y_test, predictions)))
print('RMSE : {}'.format(np.sqrt(mean_squared_error(y_test, predictions))))
```

# Построем гистограмму остатков и убедимся, что она нормально распределена.

```python
sns.displot(y_test - predictions, kde=True, bins=50);
```

```python
df_coef = pd.DataFrame(data=lr.coef_, index=X_train.columns, columns=['Coefficient'])
df_coef.sort_values(by='Coefficient', ascending=False)
```
